```python
def is_positive_definite(A):
    # Verificar se a matriz é simétrica
    if not np.allclose(A, A.T):
        return False

    # Verificar os determinantes das submatrizes principais
    n = A.shape[0]
    for i in range(1, n + 1):
        submatrix = A[:i, :i]  # Seleciona a submatriz i x i
        if np.linalg.det(submatrix) <= 0:
            return False
    return True

# Exemplo de uso com a matriz do exercício anterior
A = np.array([[12, -1, -5, 0],
              [-1, 7, 2, -1],
              [-5, 2, 10, 1],
              [0, -1, 1, 3]])

# Testar a função
resultado = is_positive_definite(A)
print(resultado)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir o intervalo de ângulos de -2π a 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Calcular os valores de seno e cosseno
y_seno = np.sin(x)
y_cosseno = np.cos(x)

# Criar o gráfico
plt.plot(x, y_seno, label='Seno')
plt.plot(x, y_cosseno, label='Cosseno')

# Adicionar títulos e legendas
plt.title('Funções Seno e Cosseno')
plt.xlabel('Ângulo (radianos)')
plt.ylabel('Valor')

# Adicionar linha horizontal no eixo y=0
plt.axhline(0, color='black', linewidth=0.5)

# Adicionar linha vertical no eixo x=0
plt.axvline(0, color='black', linewidth=0.5)

# Exibir legenda
plt.legend()

# Mostrar o gráfico
plt.grid(True)
plt.show()
```

Excercicio 4

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir a semente para reprodutibilidade
np.random.seed(1)

# Gerar a matriz X de 10x10 com valores aleatórios entre -1 e 1
X = np.random.uniform(-1, 1, (10, 10))

# Criar a matriz simétrica A = X^T * X
A = np.dot(X.T, X)

# Função para calcular o número de condicionamento
def cond_number(matrix):
    return np.linalg.cond(matrix)

# Calcular o número de condicionamento de A
cond_A = cond_number(A)
print(f"Número de condicionamento de A: {cond_A}")

# Criar uma lista de valores lambda para adicionar à matriz identidade
lambdas = np.linspace(0, 5, 100)

# Armazenar os números de condicionamento para cada lambda
cond_values = []
for lam in lambdas:
    A_prime = A + lam * np.identity(10)
    cond_values.append(cond_number(A_prime))

# Plotar o gráfico do número de condicionamento em função de lambda
plt.plot(lambdas, cond_values, label='Número de Condicionamento')
plt.xlabel('Lambda')
plt.ylabel('Número de Condicionamento')
plt.title('Número de Condicionamento de A em função de Lambda')
plt.grid(True)
plt.show()
```
